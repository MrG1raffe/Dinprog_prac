```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.linalg import sqrtm
```

```python
# Вычисляет фундаментальную матрицу X(t, .) на отрезке tau_range = [t0, t]
def fund_sol(A, tau_range):
    n = A(0).shape[0]
    ode_func = lambda y, t: y @ A(t)
    res = np.zeros((tau_range.size, n, n))
    for i in np.arange(0, n):
        y0 = np.eye(n)[i]
        res[:, i, :] = np.flip(scipy.integrate.odeint(ode_func, y0, -np.flip(tau_range)), axis = 0)
    return res
```

```python
# Возвращает центр и  матрицу конфигурации эллипсоида, дающего внешнюю оценку
# множества достижимости в момент t, точную в направлении l.
def get_ext_ellipse(param, t, l):
    sz = param['gr_sz']
    n = param['dim']
    t_range = np.linspace(param['t0'], t, sz)
    X_fund = fund_sol(param['A'], t_range)
    p0 = np.sqrt(np.dot(l, X_fund[0] @ param['X0'] @ X_fund[0].T @ l))
    p = np.zeros(sz)
    int_f1 = np.zeros((sz, n, n))
    int_f2 = np.zeros((sz, n))
    B = param['B']
    Q = param['Q']
    q = param['q']
    for i in np.arange(0, sz):
        tau = t_range[i]
        p[i] = np.sqrt(np.dot(l, X_fund[i] @ B(tau) @ Q(tau) @ B(tau).T @ X_fund[i].T @ l))
        int_f1[i] = X_fund[i] @ B(tau) @ Q(tau) @ B(tau).T @ X_fund[i].T / p[i]
        int_f2[i] = X_fund[i] @ B(tau) @ q(tau)
    center = X_fund[0] @ param['x0'] + np.trapz(int_f2, t_range, axis = 0)
    config = (p0 + np.trapz(p, t_range)) * (X_fund[0] @ param['X0'] @ X_fund[0].T / p0 + np.trapz(int_f1, t_range, axis = 0))
    return (center, config)
```

```python
# Возвращает центр и  матрицу конфигурации эллипсоида, дающего внутреннюю оценку
# множества достижимости в момент t, точную в направлении l.
def get_int_ellipse(param, t, l):
    sz = param['gr_sz']
    n = param['dim']
    t_range = np.linspace(param['t0'], t, sz)
    X_fund = fund_sol(param['A'], t_range)
    B = param['B']
    Q = param['Q']
    q = param['q']
    int_f = np.zeros((sz, n, n))
    int_fc = np.zeros((sz, n))
    for i in np.arange(0, sz):
        tau = t_range[i]
        int_fc[i] = X_fund[i] @ B(tau) @ q(tau)
        Q_sq = sqrtm(Q(tau))
        lam = np.linalg.norm(Q_sq @ B(tau).T @ X_fund[i].T @ l) / np.linalg.norm(sqrtm(param['X0']) @ X_fund[0].T @ l)
        a = Q_sq @ B(tau).T @ X_fund[i].T @ l
        b = lam * sqrtm(param['X0']) @ X_fund[0].T @ l
        Ua, da, Va = np.linalg.svd(np.reshape(a, (n, 1)))
        Ub, db, Vb = np.linalg.svd(np.reshape(b, (n, 1)))
        S = Vb / Va * Ub @ Ua.T
        int_f[i] = S @ Q_sq @ B(tau).T @ X_fund[i].T
        
    Q_st = sqrtm(param['X0']) @ X_fund[0].T + np.trapz(int_f, t_range, axis = 0)
    config = Q_st.T @ Q_st
    center = X_fund[0] @ param['x0'] + np.trapz(int_fc, t_range, axis = 0)    
    return center, config
```

```python
# Рисует двумерный эллипс с центром q и матрицей конфигурации Q по n опорным векторам
def plot_ellipse(q, Q, color = 'b', n = 1000):
    phi = np.linspace(0, 2 * np.pi, n)
    l = np.array([np.cos(phi), np.sin(phi)])
    x = Q @ l / np.sqrt(np.sum(l * (Q @ l), axis = 0))
    plt.plot(x[0] + q[0], x[1] + q[1], c = color)
```

```python
# Рисует двумерный график аппроксимации множества достижимости
# appr - матрица 2 x n
def plot_approx(appr, axis = np.array([0, 1]), color = 'b'):
    fig, ax = plt.subplots()
    fig.dpi = 70
    ax.plot(appr[0], appr[1], c = color)
    ax.set_xlabel('x' + str(axis[0] + 1))
    ax.set_ylabel('x' + str(axis[1] + 1))
    ax.grid()
    ax.axis('equal')
```

```python
# Возвращает опорный вектор по направлению l эллипсоида с матрицей конфигурации Q и центром q 
def get_sup(q, Q, l):
    return q + Q @ l / np.sqrt(np.dot(l, Q @ l))
```

```python
# Возвращает двумерную проекцию множества достижимости в момент t на оси axis
def get_approx(param, t, n = 40, mode = 'ext', axis = np.array([0, 1])):
    phi = np.linspace(0, 2 * np.pi, n)
    int_appr = np.zeros((2, n))
    ext_appr = np.zeros((2, n))
    for i in np.arange(0, n):
        l = np.eye(param['dim'])[axis[0]] * np.cos(phi[i]) + np.eye(param['dim'])[axis[1]] * np.sin(phi[i])
        if mode == 'both' or mode == 'int':
            c_int, Q_int = get_int_ellipse(param, t, l)
            int_appr[:, i] = get_sup(c_int, Q_int, l)[axis]
        if mode == 'both' or mode == 'ext':
            c_ext, Q_ext = get_ext_ellipse(param, t, l)
            ext_appr[:, i] = get_sup(c_ext, Q_ext, l)[axis]
    if mode == 'int':
        return int_appr
    if mode == 'ext':
        return ext_appr
    return (int_appr, ext_appr)
```

```python
# Вычисляет и рисует трубку достижимости в осях axis на промежутке t_range
def plot_tube(param, t_range, n_appr, axis = np.array([0, 1]), color = 'b'):
    fig = plt.figure()
    fig.dpi = 100
    ax = fig.add_subplot(111, projection='3d')
    for i in np.arange(0, t_range.size):
        x = t_range[i] * np.ones(n_appr)
        appr = get_approx(param, t_range[i], n_appr, 'ext', axis)
        y = appr[0]
        z = appr[1]
        ax.plot(x, y, z, color = color, alpha = 0.85, lw = 0.5)
        ax.set_xlabel('t')
        ax.set_ylabel('x' + str(axis[0] + 1))
        ax.set_zlabel('x' + str(axis[1] + 1))
    fig.savefig('/home/mr_giraffe/Documents/Din_prog/lab/report/tube.pdf', bbox_inches='tight')
```

```python
# Вычисляет трехмерную проекцию на оси axis множества достижимости в момент t
def get_3d_proj(param, t, n, axis = np.array([0, 1, 2])):
    phi = np.linspace(0, 2 * np.pi, 2 * n)
    psi = np.linspace(-np.pi / 2, np.pi / 2, n)
    res = np.zeros((3, 2 * n**2))
    i = 0
    x = np.zeros((2 * n, n))
    y = np.zeros((2 * n, n))
    z = np.zeros((2 * n, n))

    for (i, a1) in enumerate(phi):
        for (j, a2) in enumerate(psi):
            l = np.cos(a1) * np.cos(a2) * np.eye(param['dim'])[axis[0]] + np.sin(a1) * np.cos(a2) * np.eye(param['dim'])[axis[1]] + np.sin(a2) * np.eye(param['dim'])[axis[2]]
            c_ext, Q_ext = get_ext_ellipse(param, t, l)
            [x[i, j], y[i, j], z[i, j]] = get_sup(c_ext, Q_ext, l)[axis]
    return (x, y, z)
```

```python
# Рисует трехмерную проекцию
def plot_3d_proj(proj, axis = np.array([0, 1, 2]), color = 'purple'):
    fig = plt.figure()
    fig.dpi = 100
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(proj[0], proj[1], proj[2], alpha = 0.75)    
    ax.set_xlabel('x' + str(axis[0] + 1))
    ax.set_ylabel('x' + str(axis[1] + 1))
    ax.set_zlabel('x' + str(axis[2] + 1))
    #fig.savefig('/home/mr_giraffe/Documents/Din_prog/lab/report/set.pdf', bbox_inches='tight')
```
